```python
def split_categories(cat_list):
    if isinstance(cat_list, str):
        categories = cat_list.strip('[]').replace("'", "").split(', ')
        split_categories = []
        for cat in categories:
            split_categories.extend(cat.replace(' and ', ' & ').split(' & '))
        return [cat.strip() for cat in split_categories if cat.strip()]
    return []


folder_path = 'dataset'
books_file_path = os.path.join(folder_path, 'books_data_with_id.csv')
books_df = pd.read_csv(books_file_path, engine='python', encoding='utf-8', on_bad_lines='skip')

books_df['categories'] = books_df['categories'].apply(lambda x: split_categories(x) if isinstance(x, str) else [])

all_categories = [category for sublist in books_df['categories'] for category in sublist]

unique_categories = set(all_categories)

print("Danh mục duy nhất sau khi tách:")
for category in unique_categories:
    print(f"- {category}")
```

```python
print(len(unique_categories))
```

```python
from collections import Counter
category_counts = Counter(all_categories)
sorted_category_counts = dict(sorted(category_counts.items()))
print("Danh mục và số lần lặp lại:")
for category, count in sorted_category_counts.items():
    print(f"- {category}: {count}")
```

```python
book_id_to_display = 1

book_row = books_df[books_df['book_id'] == book_id_to_display]

if not book_row.empty:
    categories = book_row['categories'].values[0]
    print(f"Danh mục của sách với book_id {book_id_to_display}:")
    for category in categories:
        print(f"- {category}")
else:
    print(f"Sách với book_id {book_id_to_display} không được tìm thấy.")
```
